# Exercise 3A: TOP TALKERS AND LISTENERS

### Top 5 Talker & Packet send

```python
##Get the Top 5 Talker and store into dict
toptalker_count = flow_df["src_IP"].value_counts().head(5).to_dict()
```

```python
print(f"Top 5 Talker   : Number of times sending")
for key, item in toptalker_count.items():
    print(f"{key:<15}: {item}")
```

```python
#Also store the packets that the Top 5 Talker send
toptalker_pkt = toptalker_count
for key in toptalker_pkt:
    toptalker_pkt[key] = flow_df.loc[flow_df["src_IP"] == key]["packet_size"].sum()
```

```python
print(f"Top 5 Talker   : Total packet size with MAC header")
for key, item in toptalker_pkt.items():
    print(f"{key:<15}: {item}")
```

### Top 5 Listener & Packet received

```python
##Get the Top 5 Listener and store into dict
toplistener_count = flow_df["dst_IP"].value_counts().head(5).to_dict()
```

```python
print(f"Top 5 Listener: Number of times receiving")
for key, item in toplistener_count.items():
    print(f"{key:<14}: {item}")
```

```python
#Also store the packets that the Top 5 Listener received
toplistener_pkt = toplistener_count
for key in toplistener_pkt:
    toplistener_pkt[key] = flow_df.loc[flow_df["dst_IP"] == key]["packet_size"].sum()
```

```python
print(f"Top 5 Listener: Total packet size with MAC header")
for key, item in toplistener_pkt.items():
    print(f"{key:<14}: {item}")
```

# Exercise 3B: TRANSPORT PROTOCOL

```python
#TCP - 6, UDP - 17
totalTCP = (flow_df["IP_protocol"] == 6).sum()
totalUDP = (flow_df["IP_protocol"] == 17).sum()

print(f"Number of TCP protocol used: {totalTCP}")
print(f"Number of UDP protocol used: {totalUDP}")
print(f"Percentage of TCP protocol used: {totalTCP/len(flow_df)*100:.2f}%")
print(f"Percentage of UDP protocol used: {totalUDP/len(flow_df)*100:.2f}%")
```

### Percentage for all Protocol used

```python
packet_ip = []

ip_proto_df = pd.DataFrame(flow_df["IP_protocol"].value_counts())

for x in ip_proto_df.index:
    packet_ip.append(flow_df[flow_df["IP_protocol"] == x]["packet_size"].sum())

ip_proto_df["Percentage"] = (flow_df["IP_protocol"].value_counts()/len(flow_df)*100).round(decimals=2)
ip_proto_df.rename(columns={"IP_protocol": "# of Packets"},inplace=True)
ip_proto_df.insert(2, "Total Packets Size", packet_ip)
ip_proto_df.rename_axis('IP_protocol', axis=1)
```

# Exercise 3C: APPLICATION PROTOCOL

```python
app_proto_df = pd.DataFrame(flow_df["dst_port_addr"].value_counts().head(5))

packet_app = []
for x in app_proto_df.index:
    packet_app.append(flow_df[flow_df["dst_port_addr"] == x]["packet_size"].sum())
    
app_proto_df.insert(1, "Total Packets Size", packet_app)
app_proto_df.rename(columns={"dst_port_addr": "# of Packets"},inplace=True)
app_proto_df.rename_axis("Port #", axis=1)
```

Port 443 is the most frequenly used application protocol. It means most traffic passed on HTTPS.

# Exercise 3D: TRAFFIC INTENSITY

```python
#Assuming each packet is a bytes
#Assuming that SFlow traffic dataset was recorded for 15 sec.
#Packet Traffic
#Assuming sampling rate is 1/1000. Dataset is 1 sample. Hence, we multiply by a thousand
packet_traffic = flow_df["packet_size"].sum()*1000

#Convert to MB
packet_traffic_MB = packet_traffic/2**20
print(f'Packet Traffic (15s): {packet_traffic_MB:.2f}MB')

#IP Traffic
#Assuming sampling rate is 1/1000. Dataset is 1 sample. Hence, we multiply by a thousand
ip_traffic = flow_df["IP_size"].sum()*1000 

#Convert to MB
ip_traffic_MB = ip_traffic/2**20
print(f'IP Traffic (15s): {ip_traffic_MB:.2f}MB')

MAC_header = 1-ip_traffic_MB/packet_traffic_MB
print(f'Total MAC header in Packet Traffic (15s): {MAC_header*100:.2f}%')
```

### Top 5 Communication Pair

```python
ip_pair_df = pd.DataFrame(flow_df.groupby(['src_IP','dst_IP']).count()["FLOW"])
ip_pair_df = ip_pair_df.sort_values(by='FLOW',ascending=False)
ip_pair_df.columns=["Count"]
ip_pair_df.head(5)
```

Top 2 Communication Pair can be double checked with Top5Talker and Top5Listener

```python
v_df = flow_df[["src_IP","src_port_addr","dst_IP","dst_port_addr"]]
v_df["src_port_addr"] = v_df["src_port_addr"].astype(int).astype(str)
v_df["dst_port_addr"] = v_df["dst_port_addr"].astype(str)
v_df = v_df.values.tolist()
```

```python
for x in v_df:
    x[0:2] = [':'.join(x[0:2])]
    x[1:3] = [':'.join(x[1:3])]
v_df = [tuple(x) for x in v_df]
```

```python
#list for communication between hosts for graph with IP:port num
v_df
```

```python
#Building of Graph with IP pair
H = nx.Graph()
for pair in v_df:
    H.add_edge(pair[0], pair[1])
    
plt.figure(figsize=(100,100)) 
nx.draw(H, with_labels = True, node_color='black')
plt.savefig("ip_comms_1.png") # save as png
plt.show()
```

### Visualizing Communication between IP Hosts

```python
#Reusing IP Pair did on top 5 communication pair
ip_pair_list = list(ip_pair_df.index)

#Building of Graph with IP pair
G = nx.Graph()
for pair in ip_pair_list:
    G.add_edge(pair[0], pair[1])
    
plt.figure(figsize=(100,100)) 
nx.draw(G, with_labels = True, node_color='black')
plt.savefig("ip_comms.png") # save as png
plt.show()
```
